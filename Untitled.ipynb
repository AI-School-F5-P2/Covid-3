```python
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import requests

# Obtener los datos de Covid19 desde Our World in Data
def get_covid_data():
    url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
    response = requests.get(url)
    if response.status_code == 200:
        return pd.read_csv(url)
    else:
        print("Error al obtener los datos. Código de estado:", response.status_code)
        return None

# Convertir datos en un DataFrame
df = get_covid_data()

# Asegúrate de inspeccionar la estructura de 'df' y ajustar el código según los datos disponibles
# Por ejemplo, puedes utilizar 'df.info()' para obtener información sobre las columnas y tipos de datos

# Visualización con Matplotlib
plt.figure(figsize=(12, 6))
countries = ['USA', 'Canada', 'Brazil', 'India', 'United Kingdom']
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.xlabel('Fecha')
plt.ylabel('Nuevos Casos Diarios')
plt.legend(loc='upper left')
plt.title('Nuevos Casos Diarios de Covid19 en Países Seleccionados')

# Visualización con Bokeh
output_notebook()
p = figure(x_axis_type="datetime", plot_width=800, plot_height=400, title="Nuevos Casos Diarios de Covid19 en Países Seleccionados")
for country in countries:
    country_data = df[df['location'] == country]
    p.line(country_data['date'], country_data['new_cases'], line_width=2, legend_label=country)

p.legend.location = "top_left"
p.legend.click_policy = "hide"

show(p)

# Análisis y conclusiones (realiza tu análisis aquí)

```
