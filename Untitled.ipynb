```python
import requests
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np

#link del dathaset
url = "https://api.covidtracking.com/v1/states/daily.json"
```

```python
# Realizar la solicitud (request) para obtener los datos JSON
response = requests.get(url)

# Verificar que la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Cargar los datos JSON en un DataFrame de pandas 
    data = response.json()
    df = pd.DataFrame(data)
    
    # Mostrar las primeras filas del DataFrame
    display(df)
else:
    print("No se pudo obtener los datos JSON. Código de estado:", response.status_code)
```

```python
def mostrar_resumen_casos(df):
    # Filtrar las columnas 'positive' y 'probableCases'
    casos_positivos = df['positive']
    casos_probables = df['probableCases']
    
    # Calcular la suma total de casos positivos y probables
    total_casos_positivos = casos_positivos.sum()
    total_casos_probables = casos_probables.sum()
    
    # Calcular el promedio de casos positivos y probables
    promedio_casos_positivos = casos_positivos.mean()
    promedio_casos_probables = casos_probables.mean()
    
    # Calcular el máximo y mínimo de casos positivos y probables
    max_casos_positivos = casos_positivos.max()
    min_casos_positivos = casos_positivos.min()
    max_casos_probables = casos_probables.max()
    min_casos_probables = casos_probables.min()
    
    # Mostrar los resultados
    print("Resumen de casos positivos:")
    print("Total de casos positivos:", total_casos_positivos)
    print("Promedio de casos positivos:", promedio_casos_positivos)
    print("Máximo de casos positivos:", max_casos_positivos)
    print("Mínimo de casos positivos:", min_casos_positivos)
    
    print("\nResumen de casos probables:")
    print("Total de casos probables:", total_casos_probables)
    print("Promedio de casos probables:", promedio_casos_probables)
    print("Máximo de casos probables:", max_casos_probables)
    print("Mínimo de casos probables:", min_casos_probables)

# Llamar a la función con el DataFrame 'df'
mostrar_resumen_casos(df)
```

```python
def grafica_casos_positivos(df):
    # Filtrar las columnas 'date' y 'positive'
    fechas = df['date']
    casos_positivos = df['positive']
    
    # Crear la gráfica de barras
    plt.figure(figsize=(12, 6))
    plt.bar(fechas, casos_positivos, color='skyblue')
    plt.xlabel('Fecha')
    plt.ylabel('Casos Positivos')
    plt.title('Evolución de Casos Positivos de COVID-19')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Llamar a la función con el DataFrame 'df'
grafica_casos_positivos(df)
```

```python
def analisis_casos_negativos(df):
    # Filtrar la columna 'negative'
    casos_negativos = df['negative']
    
    # Calcular la suma total de casos negativos
    total_casos_negativos = casos_negativos.sum()
    
    # Calcular el promedio de casos negativos
    promedio_casos_negativos = casos_negativos.mean()
    
    # Calcular el máximo y mínimo de casos negativos
    max_casos_negativos = casos_negativos.max()
    min_casos_negativos = casos_negativos.min()
    
    # Calcular el número de días con resultados negativos
    dias_con_negativos = casos_negativos.count()
    
    # Mostrar los resultados
    print("Resumen de casos negativos:")
    print("Total de casos negativos:", total_casos_negativos)
    print("Promedio de casos negativos:", promedio_casos_negativos)
    print("Máximo de casos negativos:", max_casos_negativos)
    print("Mínimo de casos negativos:", min_casos_negativos)
    print("Número de días con resultados negativos:", dias_con_negativos)

# Llamar a la función con el DataFrame 'df'
analisis_casos_negativos(df)
```

```python
def grafica_casos_negativos(df):
    # Filtrar las columnas 'date' y 'negative'
    fechas = df['date']
    casos_negativos = df['negative']
    
    # Crear la gráfica de barras
    plt.figure(figsize=(12, 6))
    plt.bar(fechas, casos_negativos, color='lightcoral')
    plt.xlabel('Fecha')
    plt.ylabel('Casos Negativos')
    plt.title('Evolución de Casos Negativos de COVID-19')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Llamar a la función con el DataFrame 'df'
grafica_casos_negativos(df)
```
